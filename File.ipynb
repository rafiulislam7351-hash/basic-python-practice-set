```python
import random

def  game():
    print("The game is starting")
    score=random.randint(1,100)
    with open("hiscore.txt") as f:
        hiscore=f.read()
        if hiscore!="":
            hiscore=int(hiscore)
        else:
            hiscore=0
    print(f"Your score:{score}")
    if score>hiscore:
        with open("hiscore.txt","w") as f:
            f.write(str(score))
    return score
game()
```

```python
def multiplication_table(n):
    table=""
    for i in range(1,11):
        table+=f"{n} x {i} = {n*i}\n"
    with open("hiscore.txt","a") as f:
        f.write(table)
for i in range(2,21):
    multiplication_table(i)
```

```python
word="donkey"
with open("hiscore.txt","r") as f:
    content=f.read()
new_content=content.replace(word,"######")
with open("hiscore.txt","w") as f:
    f.write(new_content)
```

```python
words=["donkey","pinky","smelly"]
with open("hiscore.txt","r") as f:
    content=f.read()
for word in words:
    content=content.replace(word,"#"*len(word))
with open("hiscore.txt","w") as f:
    f.write(content)
```

```python
with open("log.txt","r") as f:
    content=f.read()
if("python" in content):
    print("Yes! Python is present")
else:
    print("No python is not present")
```

```python
with open("log.txt","r") as f:
    lines=f.readlines()
line_number=1
for line in lines:
    if("python" in line):
        print(f"Yes! Python is present in line number:{line_number}")
        break
    line_number+=1
else:
    print("No python is not present")
```
