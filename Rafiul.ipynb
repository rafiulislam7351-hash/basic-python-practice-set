```python
import random
import os

# --- STEP 1: READ THE SCORE ---
# We check if the file exists first so the program doesn't crash
if os.path.exists("scores.txt"):
    with open("scores.txt", "r") as file:
        score = int(file.read()) # Convert the text in the file to a number
else:
    score = 0 # Start at 0 if no file exists

print(f"Current Score: {score}")

# --- STEP 2: PLAY THE GAME ---
computer = -random.choice([1, -1, 0])
player_input = input("Enter your choice (Rock, Paper, Scissors): ").capitalize()

playerDict = {"Rock": 1, "Paper": -1, "Scissors": 0}
computerDict = {1: "Rock", -1: "Paper", 0: "Scissors"}

if player_input in playerDict:
    print("Computer's choice =", computerDict[computer])
    player_num = playerDict[player_input]

    # Logic to update score
    if (computer == -1 and player_num == 0) or \
       (computer == 1 and player_num == -1) or \
       (computer == 0 and player_num == 1):
        print("You Win!")
        score += 1  # Add 1 to score
    elif computer == player_num:
        print("Draw (>_<)")
    else:
        print("You Lose!")
        score -= 1  # Subtract 1 from score

    # --- STEP 3: WRITE THE NEW SCORE ---
    with open("scores.txt", "w") as file:
        file.write(str(score)) # Convert number to string to save it

    print(f"New Score Saved: {score}")

else:
    print("Invalid input!")
import random
import os

# --- STEP 1: READ THE SCORE ---
# We check if the file exists first so the program doesn't crash
if os.path.exists("scores.txt"):
    with open("scores.txt", "r") as file:
        score = int(file.read()) # Convert the text in the file to a number
else:
    score = 0 # Start at 0 if no file exists

print(f"Current Score: {score}")

# --- STEP 2: PLAY THE GAME ---
computer = -random.choice([1, -1, 0])
player_input = input("Enter your choice (Rock, Paper, Scissors): ").capitalize()

playerDict = {"Rock": 1, "Paper": -1, "Scissors": 0}
computerDict = {1: "Rock", -1: "Paper", 0: "Scissors"}

if player_input in playerDict:
    print("Computer's choice =", computerDict[computer])
    player_num = playerDict[player_input]

    # Logic to update score
    if (computer == -1 and player_num == 0) or \
       (computer == 1 and player_num == -1) or \
       (computer == 0 and player_num == 1):
        print("You Win!")
        score += 1  # Add 1 to score
    elif computer == player_num:
        print("Draw (>_<)")
    else:
        print("You Lose!")
        score -= 1  # Subtract 1 from score

    # --- STEP 3: WRITE THE NEW SCORE ---
    with open("scores.txt", "w") as file:
        file.write(str(score)) # Convert number to string to save it

    print(f"New Score Saved: {score}")

else:
    print("Invalid input!")
```
